```python
with open('inputs/14.txt', 'r') as file:
    inp = file.read()[:-1]
```

```python
import numpy as np
```

```python
formula, *rules = inp.split("\n")
rules = rules[1:]
```

```python
ins = {}
for r in rules:
    ins[r[:2]] = r[-1]
```

```python
def combine(a,b):
    c = ""
    for i in range(len(a)):
        if i != 0:
            c += b[i-1]
        c += a[i]
    return c
```

```python
def chemistry(formula, steps=10): 
    for step in range(steps):
        new = ""
        for i in range(len(formula)-1):
            new += ins[formula[i:i+2]]
        formula = combine(formula, new)
    return formula
```

```python
# 1
letters = chemistry(formula, 10)
count = {}
for i in letters:
    if i not in count: count[i] = 1
    else: count[i] += 1
max(list(count.values())) - min(list(count.values()))
```

```python
letters = chemistry(formula, 20)
count = {}
for i in letters:
    if i not in count: count[i] = 1
    else: count[i] += 1
print(count)
```

```python
count_after_20 = {}
for rule in ins:
    letters = chemistry(rule, 20)
    count = {}
    for i in letters:
        if i not in count: count[i] = 1
        else: count[i] += 1
    count_after_20[rule] = count
count_after_20
```

```python
from collections import defaultdict
```

```python
letters = chemistry(formula, 20)

count = defaultdict(int)
for i in range(len(letters)-1):
    values = count_after_20[letters[i:i+2]]
    for v in values:
        count[v] += values[v]
        
letter_count = defaultdict(int)
for i in letters[1:-1]:
    letter_count[i] += 1
    
for v in letter_count:
    count[v] -= letter_count[v]
    
count
```

```python
# 2
values = []
for i in count:
    values.append(count[i])
print(max(values)-min(values))
```

```python
from collections import Counter                             # Saves time of converting lists to dicts

tpl, _, *rules = inp.split('\n')                            # Use ,_, to get rid of the empty thing in between
rules = dict(r.split(" -> ") for r in rules)                # Can convert list in list to dict!
pairs = Counter(map(str.__add__, tpl, tpl[1:]))             # Is much better than iterating through everything to get dict
chars = Counter(tpl)                                        

for _ in range(40):
    for (a,b), c in pairs.copy().items():                   # copy() so dict doesn't change size during operation, items() gives keys and values
        x = rules[a+b]                                      # new letter depending on keys
        pairs[a+b] -= c                                     # "delete" old pair 
        pairs[a+x] += c                                     # spawn 2 new pairs from old
        pairs[x+b] += c                       
        chars[x] += c                                       # increase characters by value
print(max(chars.values())-min(chars.values()))              # .values() gives only values of dict
```
